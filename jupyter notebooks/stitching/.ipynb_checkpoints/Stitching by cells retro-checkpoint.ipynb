```python
"""Functions for plotting tca decomp."""
import os
import flow
import pool
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import tensortools as tt
import seaborn as sns
import pandas as pd
from copy import deepcopy
from cascade import df
from cascade import tca
# from random import shuffle
from cascade import paths
```

### set up params for TCA loading and DateSorter

```python
mouse='OA27'
trace_type='zscore_day'
method='ncp_bcd'
cs=''
warp=False
word='ratios'

drive_css=('0', '135', '270'),
```

```python
pars = {'trace_type': trace_type, 'cs': cs, 'warp': warp}
days = flow.DateSorter.frommeta(mice=[mouse], tags=None)
```

```python
rev_days = [i for i in days[::-1]]
```

### get all of your variables across days for learning: pre-reversal --> naive

```python
# loop through your datesorter backwards
rank_num = 10
conds_by_day = []
trialerr_by_day = []
neuron_ids_by_day = []
neuron_clusters_by_day = []
factors_by_day = []
for day1 in days[::-1]:
# day1 = days[0]

    # load dir
    load_dir = paths.tca_path(mouse, 'single', pars=pars, word=word)
    tensor_path = os.path.join(load_dir, str(day1.mouse) + '_' + str(day1.date)
                               + '_single_decomp_' + str(trace_type) + '.npy')
    input_tensor_path = os.path.join(load_dir, str(day1.mouse) + '_'
                                     + str(day1.date)  + '_single_tensor_' + str(trace_type)
                                     + '.npy')
    input_ids_path = os.path.join(load_dir, str(day1.mouse) + '_'
                                  + str(day1.date) + '_single_ids_'
                                  + str(trace_type) + '.npy')
    meta_path = os.path.join(load_dir, str(day1.mouse) + '_' + str(day1.date)
                         + '_df_single_meta.pkl')
    
    # load your metadata, skip post reversal days
    meta = pd.read_pickle(meta_path)
    condition = meta['condition']
    trialerror = meta['trialerror']
    if 'reversal1' in meta['learning_state'].unique() or 'reversal2' in meta['learning_state'].unique():
        continue
    # skip days that do not have minus, AND neutral
    if 'minus' not in meta['condition'].unique() \
    or 'neutral' not in meta['condition'].unique() \
    or 'plus' not in meta['condition'].unique():
        continue
    print(day1)
    
    # load your data
    ensemble = np.load(tensor_path)
    ensemble = ensemble.item()
    ids = np.load(input_ids_path)

    # sort neuron factors by component they belong to most
    sort_ensemble, my_sorts = tca._sortfactors(ensemble[method])
    
#     d1_drive = []
#     for dcs in drive_css:
#         try:
#             d1_drive.append(pool.calc.driven.trial(day1, dcs))
#             print(pool.calc.driven.trial(day1, dcs))
#         except KeyError:
#             print(str(day1) + ' requested ' + dcs +
#                   ': no match to what was shown (probably pav only).')

# cell_ids = flow.xday._read_crossday_ids(day.mouse, day.date)

    cell_ids = {} # keys are rank
    cell_clusters = {}
    itr_num = 0  # use only best iteration
    for k in sort_ensemble.results.keys():
        factors = sort_ensemble.results[k][itr_num].factors[0]
        # these are already sorted, so these will define clusters, no need to sort again
        max_fac = np.argmax(factors, axis=1)
        cell_clusters[k] = max_fac
        cell_ids[k] = ids[my_sorts[k-1]]
    
    k = rank_num
#     cell_ensembles[k]
    neuron_ids_by_day.append(cell_ids[k])
    neuron_clusters_by_day.append(cell_clusters[k])
    factors_by_day.append(sort_ensemble.results[k][0])
    conds_by_day.append(condition)
    trialerr_by_day.append(trialerror)
```

### match cell ensembles using highest weights of same cells on previous day

```python
plus_only = False
temp_fac_map = np.zeros((np.shape(factors_by_day[0].factors[1])[1],
                   np.shape(factors_by_day[0].factors[1])[0],
                   len(factors_by_day)))
tri = []
tri_conds = []
for clus_val in np.unique(neuron_clusters_by_day[0]):
    
    # set seed cluster
    starting_clus = clus_val
    trial_fac_map = []
    
    # set day counter
    d1 = 0
    
    # add in your starting temporal factor for a cluster
    if plus_only:
        plus_weights = (conds_by_day[0] == 'plus')
        plus_scalar = np.nanmean(factors_by_day[0].factors[2][plus_weights,clus_val])
        temp_fac_map[clus_val,:,0] = factors_by_day[0].factors[1][:,clus_val]*plus_scalar
        trial_fac_map.append(factors_by_day[0].factors[2][:,clus_val])
    else:
        temp_fac_map[clus_val,:,0] = factors_by_day[0].factors[1][:,clus_val]
        trial_fac_map.append(factors_by_day[0].factors[2][:,clus_val])
    
    for d2 in range(1,len(neuron_clusters_by_day)):
        
        # get cell clusters (highest neuron factor weight) for pair of subsequent days
        clus1 = neuron_clusters_by_day[d1]
        clus2 = neuron_clusters_by_day[d2]
        
        # get cell ids for pair of subsequent days
        ids1 = neuron_ids_by_day[d1]
        ids2 = neuron_ids_by_day[d2]
        
        # get neuron weights for the second day
        neur_fac2 = factors_by_day[d2].factors[0]
        
        # which ids are in your starting cluster?
        id_cluster = ids1[clus1 == starting_clus]
        
        # do cells from this cluster match on the previous day (rev order)
        new_clusters = np.unique(clus2[np.isin(ids2, id_cluster)])
        # skip this day-pair if cells are not matched to any cluster
        if len(new_clusters) == 0:
            # on skipped days blank trials (set to zeros)
            trial_fac_map.append(np.zeros(len(factors_by_day[d2].factors[2][:,0])).flatten())
            continue
        # find cluster on subsequent day with highest summed cell weights   
        clus_weight = []
        for i in new_clusters:
            bool_clus = np.isin(ids2, id_cluster) & np.isin(clus2, i)
            if plus_only:
                plus_weights = (conds_by_day[d2] == 'plus')
                plus_scalar = np.nanmean(factors_by_day[d2].factors[2][plus_weights,i])
                clus_weight.append(np.sum(neur_fac2[bool_clus,i]*plus_scalar))
            else:
                clus_weight.append(np.sum(neur_fac2[bool_clus,i]))
        best_clus = new_clusters[np.argmax(clus_weight)]
        
        # get temporal/trial factors for the best cluster
        temp_fac_map[clus_val,:,d2] = factors_by_day[d2].factors[1][:,best_clus] 
        trial_fac_map.append(factors_by_day[d2].factors[2][:,best_clus])
        
        # start your next search from the best matched cluster
        starting_clus = best_clus
        d1 = d2
                      
    # append cluster trial factors together
    tri.append(trial_fac_map)
```

## updated matching algo, all pairwise squid

```python
# get all pairwise correlation coeffs for matched neuron factors 
# this should be done on a time forward vector and can include reversal

# Do forward and reverse matching and just divide by double,
# we are trying to get all edge weights
# neuron_ids_by_day = neuron_ids_by_day
# factors_by_day = factors_by_day[::-1]  ####### THESE TWO ARE IN THE FORWARD DIRECTION RIGHT NOW
sim_mats_by_day = []
for k in range(len(factors_by_day)):
    
    # sim mats, day k comps are columns, day i comps are rows 
    sim_mats = np.zeros((rank_num, rank_num, len(factors_by_day)))
    
    for i in range(len(factors_by_day)):
    
        # always compare in the time-forward direction (remember that factors_by_day is in reverse order)
        ids1 = neuron_ids_by_day[k]
        ids2 = neuron_ids_by_day[i]
        ids1_bool = np.isin(ids1, ids2)
        ids2_bool = np.isin(ids2, ids1)

        ids1_sort = np.argsort(ids1[ids1_bool])
        ids2_sort = np.argsort(ids2[ids2_bool])

        ids1_weights = factors_by_day[k].factors[0][ids1_bool,:]
        ids2_weights = factors_by_day[i].factors[0][ids2_bool,:]

        # compare correlations of vectors that have been sorted to match neuron orders 
        ids_corr = np.corrcoef(ids1_weights[ids1_sort,:].T, y=ids2_weights[ids2_sort,:].T)
        sim_mats[:,:,i] = deepcopy(ids_corr[-10:, 0:10])

    sim_mats_by_day.append(sim_mats)
```

```python
# "Squid matching"
# get a transition matrix for consecutive days in the forward direction
# i.e. for rank 10 tca: 10x10 matrix where each A[:,i] are the counts that a factor was matched to 
# output is a set of 10x10 matrices for each paiir of days (so length is 1 less than day number)

trans_mats = np.zeros((rank_num, rank_num, len(factors_by_day)-1))
for i in range(len(sim_mats_by_day)-1):
    
    # FORWARD day[i]-->day[i+1]
    # best match of day i components to day n
    step_outward_lookup = np.argmax(sim_mats_by_day[i], axis=0)
    # best match of day n components to day i
    step_inward_lookup = np.argmax(sim_mats_by_day[i+1], axis=1) ## DOUBLE CHECK AXIS IS CORRECT
    
    # get transition matrix A[:,k], where k is comp number on day i and A[j,k] is probability
    # of compenent k being matched to comp j on SUBSEQUENT day
    counter_mat = np.zeros((rank_num, rank_num))
    for k in range(rank_num):
        vec_out = step_outward_lookup[k,:]
        for j in range(len(vec_out)):
            best_step_inward = step_inward_lookup[vec_out[j],j]
            counter_mat[best_step_inward,k] += 1
    
    # aggregate results for all paired-day comparisons
    trans_mats[:,:,i] = counter_mat
    
    # ---------------------------------------------------------------------------------
    
    # REVERSE day[i]<--day[i+1]
    # best match of day n components to day i (the opposite comparison as FORWARD block)
    step_outward_lookup2 = np.argmax(sim_mats_by_day[i+1], axis=0)
    # best match of day i components to day n
    step_inward_lookup2 = np.argmax(sim_mats_by_day[i], axis=1) ## DOUBLE CHECK AXIS IS CORRECT

    # get transition matrix A[:,k], where k is comp number on day n (i+1) and A[j,k] is probability
    # of compenent k being matched to comp j on PREVIOUS day
    counter_mat2 = np.zeros((rank_num, rank_num))
    for k in range(rank_num):
        vec_out2 = step_outward_lookup2[k,:]
        for j in range(len(vec_out2)):
            best_step_inward2 = step_inward_lookup2[vec_out2[j],j]
            counter_mat2[best_step_inward2,k] += 1
    
    # transpose and add to counter_mat results, this should now be in the same reference frame
    trans_mats[:,:,i] += counter_mat2.T
    
# if you now divide this by rank_num*2 (number of comparisons) you 
# should get a "probability" of a given transition
trans_mats = trans_mats/(rank_num*2)
```

```python
# "Path of least resistance"
# step through transition matrix and match factors
# build a single transition matrix for each factor/component
# for k in range(rank_num):
k = 0
comp_idx = k
best_path = np.zeros((np.shape(trans_mats)[2]))
best_prob = np.zeros((np.shape(trans_mats)[2]))
for i in range(np.shape(trans_mats)[2]):
    comp_idx_new = np.argmax(trans_mats[:, comp_idx, i])
    best_path[i] = comp_idx_new
    best_prob[i] = trans_mats[comp_idx_new, comp_idx, i]
    comp_idx = comp_idx_new
```

```python
# new way of aligning: zip through 
# get factor index stepping through similarity matrices for pairs of days


match_by = 'combo_sim' # similarity, probability, combo

# choose list of simialrity matrices to use for matching components
# across days
if match_by == 'similarity':
    match_mat = sim_mat_by_day
elif match_by == 'probability':
    match_mat = prob_mat_by_day
elif match_by == 'combination': # elementwise multiplication of the prob mat 
    match_mat = [np.multiply(a,b) for a,b in zip(sim_mat_by_day, prob_mat_by_day)]
elif match_by == 'combo_sim': # elementwise multiplication of the prob mat 
    match_mat = [np.multiply(a,b,c) for a,b,c in zip(sim_mat_by_day, prob_mat_by_day, sim_mat_tempo_by_day)]
else:
    print('Unregognized matching method in match_by.')
#     break

temporal_factors_list = []
transition_weights = np.zeros((np.shape(match_mat[0])[0], len(match_mat)))
# factors_by_day_ = factors_by_day[::-1]
for comp_num in range(np.shape(match_mat[0])[0]):
    
    temporal_factors = []
    for i in range(len(match_mat)):

        # get similarity matrices. (i.e., [day1-->day2]) reverse order matching, forward in time.
#         sim_mat = match_mat[::-1][i]
        sim_mat = match_mat[i]
        
        if i == 0:
            starting_comp = np.argmax(sim_mat[comp_num, :])
            transition_weights[comp_num, i] = sim_mat[comp_num, starting_comp]
            fac_1 = factors_by_day[i+1].factors[1][:, starting_comp] # factors_by_day already in reverse order
            fac_2 = factors_by_day[i].factors[1][:, comp_num]
            temporal_factors.append(fac_2)
            temporal_factors.append(fac_1)
        else:
            new_starting_comp = np.argmax(sim_mat[starting_comp, :])
            transition_weights[comp_num, i] = sim_mat[starting_comp, new_starting_comp]
            fac_n = factors_by_day[i+1].factors[1][:,new_starting_comp]
            temporal_factors.append(fac_n)
            starting_comp = new_starting_comp
            
    temporal_factors_list.append(temporal_factors) 
        
```
